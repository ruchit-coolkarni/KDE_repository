```python
# Group by year and then apply the quintile count calculation for each group
quintile_counts_by_year = filtered_df.groupby('Year').apply(lambda df: pd.Series({
    'quintile_1_count': df[(df['Rainfall..mm.'] >= quintile_1_lower) & (df['Rainfall..mm.'] <= quintile_1_upper)].shape[0],
    'quintile_2_count': df[(df['Rainfall..mm.'] > quintile_1_upper) & (df['Rainfall..mm.'] <= quintile_2_upper)].shape[0],
    'quintile_3_count': df[(df['Rainfall..mm.'] > quintile_2_upper) & (df['Rainfall..mm.'] <= quintile_3_upper)].shape[0],
    'quintile_4_count': df[(df['Rainfall..mm.'] > quintile_3_upper) & (df['Rainfall..mm.'] <= quintile_4_upper)].shape[0],
    'quintile_5_count': df[(df['Rainfall..mm.'] > quintile_4_upper) & (df['Rainfall..mm.'] <= quintile_5_upper)].shape[0]
}))

print(quintile_counts_by_year)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Assuming quintile_counts_by_year is a DataFrame where each column represents a quintile count over years
# and the index of DataFrame is the years.

# Iterate over each quintile column to plot and analyze trends
for quintile in quintile_counts_by_year.columns:
    # Extract year and counts for the current quintile
    years = quintile_counts_by_year.index
    counts = quintile_counts_by_year[quintile]
    
    # Linear regression for trend analysis
    slope, intercept, r_value, p_value, std_err = linregress(years, counts)
    trend_line = slope * years + intercept  # Calculate the trend line
    
    # Plotting
    plt.figure(figsize=(10, 5))
    plt.plot(years, counts, label=f'Counts for {quintile}', marker='o')  # Plot the actual counts
    plt.plot(years, trend_line, 'r--', label=f'Trend line (slope={slope:.2f})')  # Plot the trend line
    
    # Add labels and title
    plt.title(f'Trend Analysis for {quintile}')
    plt.xlabel('Year')
    plt.ylabel('Count')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Print the trend analysis results
    trend_direction = 'increasing' if slope > 0 else 'decreasing' if slope < 0 else 'insignificant'
    print(f"The trend for {quintile} is {trend_direction}. Slope: {slope:.2f}, p-value: {p_value:.4f}\n")
```

```python
import matplotlib.pyplot as plt
from scipy.stats import linregress
import pandas as pd

# Load the dataset
df = pd.read_csv('extreme_rainfall_events.csv')

# Count the number of extreme rainfall events per year
extreme_rainfall_count = df.groupby('Year').size()

# Plot the scatter of extreme events per year
plt.figure(figsize=(10, 6))
plt.scatter(extreme_rainfall_count.index, extreme_rainfall_count.values, color='b', label='Extreme Events')

# Perform linear regression to analyze the trend
slope, intercept, r_value, p_value, std_err = linregress(extreme_rainfall_count.index, extreme_rainfall_count.values)

# Add trend line to the scatter plot
plt.plot(extreme_rainfall_count.index, intercept + slope * extreme_rainfall_count.index, 'r', label=f'Trend line (slope={slope:.2f})')

# Enhance plot
plt.title('Count of Extreme Rainfall Events (>30mm) Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Extreme Rainfall Events')
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()

# Print the slope and p-value to interpret the trend statistically
print(f"Slope of trend line: {slope:.2f}")
print(f"P-value of the regression: {p_value:.4f}")

# Interpretation based on slope and p-value
if p_value < 0.05:  # Assuming a significance level of 5%
    if slope > 0:
        print("Statistically significant increase in extreme rainfall events.")
    elif slope < 0:
        print("Statistically significant decrease in extreme rainfall events.")
else:
    print("No statistically significant trend in extreme rainfall events.")
```
